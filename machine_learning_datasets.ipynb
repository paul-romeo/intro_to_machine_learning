```python
import mglearn
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

# Some datasets to be used in The Intro to Machine Learning

## 1. The "make_forge" dataset from the mglearn library 

```python

# mglearn: make_forget datasets
X, y = mglearn.datasets.make_forge()

# Plotting make_forge datasets 
mglearn.discrete_scatter(X[:,0], X[:,1], y)

plt.legend(["class 0", "class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X shape:{}\n".format(X.shape))
```

## 2. The "make_wave" dataset  from the mglearn library 

```python
X, y = mglearn.datasets.make_wave()

# Plotting make_wave datasets 
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("Feature")
plt.ylabel("Target regression")    # regression label 
```

## 3. The Wisconsin breast cancer dataset from the scikit-learn library
###  This dataset contains the measurement on the patient tumors. Tumors can be categorized into two kinds: 
<li>benign: harmless tumour 
<li>malignant: cancerous tumour

Notes: 
<li>scikit-learn datasets are usually stored in the Bunch objects. Bunch objects is accessed by using the dot (.) notation 
<li>load_breast_cancer function is used to load the breast cancer datasets 

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

# Display the field (column) names of the datasets 
print("cancer keys():\n{}".format(cancer.keys()))
```

```python
# Display the cancer description
print("Description:\n{}".format(cancer.DESCR))
```

```python
# Display the number of data points (aka number of records)
#   and the number of features on each data point 
print("Shape of the cancer data:{}".format(cancer.data.shape))
```

```python
# Display the sample counts of each class (malignant and benign)
print("Sample counts per class:\n{}".format({n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}))
```

```python
# Display the feature names (aka parameters)
print("Feature names:\n{}".format(cancer.feature_names))
```

## 4. The Boston Housing from the scikit-learn library 
#### Description: This is a real-world regression dataset. The task associated with this dataset is to predict the median value of homes in several Boston neighborhoods in the 1970s, using information such as crime rate, proximity to the Charles River, highway accessibility, and so on. The dataset contains 506 data points, described by 13 features

```python
from sklearn.datasets import load_boston

boston = load_boston()
print("boston.keys():\n{}".format(boston.keys()))
```

```python
# Display the Boston DESCRIPTION
print("Boston Housing Description:\n{}".format(boston.DESCR))
```

```python
# Display the Boston data shape. REsult: 506 records with 13 features per record 
print("Boston.data.shape:\n{}".format(boston.data.shape))
```

```python
# Display the Boston feature names 
print("Boston feature names:\n{}".format(boston.feature_names))
```
